# Slicing & Subsetting with .loc & .iloc:

```python
#Q1. Set the weather index as "MinTemp" and "MaxTemp" then sort it.

weather.set_index(['MinTemp','MaxTemp']).sort_index()
```

```python
#Slice outer index level

#Q2. Slice the MinTemp from (6.1 to 8.0)

weather.set_index(['MinTemp','MaxTemp']).query('MinTemp>=6.1 and MinTemp<=8.0')
```

```python
#Slice the inner index level

#Q3. Slice the first four index i.e, (6.1,18.2):(8.0,24.3)

weather.set_index(['MinTemp','MaxTemp']).loc[([6.1,8.0],[18.2,24.3]),:]
```

```python
#Slicing columns

#Q4. Slice weather by columns (Rainfall and WindGustSpeed)

weather[['Rainfall','WindGustSpeed']]
```

```python
#Q5. Slice weather twice from (6.1, 18.3 to 8.0,24.3) keeping columns RainToday and RainTomorrow

display(weather.query('MinTemp in (6.1,8.0) and MaxTemp in (18.3,24.3)')) 
#display is used when you want output of two commands within one cell or
weather.loc[(weather['MinTemp'].isin([6.1,8.0])) & (weather['MaxTemp'].isin([18.3,24.3])),:]
```

```python
#Slicing by dates

#Q6. Set "Date" as index , sort() it and Slice date between 2020-01-01 and 2020-04-10

weather.set_index('Date').sort_index().query('Date>="2020-01-01" and Date<="2020-04-10"')
```

```python
#Subsetting by rows and columns

#Q7. Subset weather by row (2 to 5) and columns (1 to 4)
weather.iloc[2:6,1:5]
```

# Working with pivot tables:

```python
#Q1. Pivot the weather by keeping value as "Rainfall", index as "MinTemp" and columns as "MaxTemp".

weather.pivot_table(index='MinTemp',columns='MaxTemp',values='Rainfall')
```

```python
#Q2. Slice first five index

weather.pivot_table(index='MinTemp',columns='MaxTemp',values='Rainfall').iloc[:5]
```

```python
#Q3. Calculate summary stats across index and columns separately, i.e (mean)
weather.pivot_table(index='MinTemp',columns='MaxTemp',values='Rainfall',aggfunc='mean')
```

```python
#import matplotlib and pandas 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the weather data
weather1=pd.read_csv("C:\Data science\Machine learning with phyton\data set\Assignmnet(Slicing and Visualizing)\weather.csv")
```

```python
weather1.head()
```

```python
#Q1. Plot histogram of "Rainfall" keeping bins as 25 and show it.

weather1['Rainfall'].plot.hist(bins=25)
```

```python
#Bar plots

#Q2. Assign "Rainfall" & "RainToday" to df and perform bar plot.

df=sns.barplot(x='RainToday',y='Rainfall',data=weather1)
plt.show()
```

```python
#Line plot

#Q3. Assign "RainToday" and "WindGustSpeed" to df1 and perform line plot

df1=sns.lineplot(x='RainToday',y="WindGustSpeed",data=weather1)
plt.show()
```

```python
#Layering plots

#Q4. Layer the plots keeping the dataframe as "RainTomorrow" & "Rainfall" and also add a legend("Yes","No")

df2=weather1[['RainTomorrow','Rainfall']]
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='Yes', label='Yes')
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='No', label='No')
plt.legend()
plt.show()
```

```python
#Q5. Add Transperancy , i.e alpha of 0.5

df2=weather1[['RainTomorrow','Rainfall']]
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='Yes', label='Yes',alpha=0.5)
plt.bar(df2['Rainfall'],df2['RainTomorrow']=='No', label='No',alpha=0.5)
plt.legend()
plt.show()
```
