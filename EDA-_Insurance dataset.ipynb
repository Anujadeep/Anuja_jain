```python
# splitting the dataset into the continous and categorical variable:
cat=['sex','smoker','region']
con=['age','bmi','children','charges']
```

# univariate analysis:

```python
for i in cat:
    sns.countplot(i,data=df)
    plt.show()
```

# observation:
1. Number of males and females are equal around 650.
2. Most of the people around 1000 dont smoke and only about 250 people smoke.
3. 350 are from Southeast and the remaining southwest,northwest,northeast are 300.


```python
for i in con:
    sns.distplot(df[i])
    plt.show()
```

# observation:
1. bmi are normally distributed.
2. charges and chidern are negative linear relationship.

# bivariate analysis:

```python
for i in cat:
    sns.countplot(x=i,hue='sex',data=df)
    plt.show()
```

# observation:
1. Around 500 people both(male and female) do not smoke and only 100 female and 150 male smoke.
2. Southeast people are about 170(female) and 180(male) rest all (southwest,northwest,northeast) people are around 160(female      and male).

```python
for i in cat:
    sns.swarmplot(x='charges',y=i,data=df)
    plt.show()
```

# observation:
1. Both male and female both have a more charges in between 0 to 1500.
2. People who have a charges in between 0 to 1500 dont smoke and those who have a charges
   in between 1700 to 5000 they smoke.
3. All region people have a more charges in between 0 to 1500
